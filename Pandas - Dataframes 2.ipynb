```python
import pandas as pd
import numpy as np
from numpy.random import randn
```

## Conditional Selection

```python
df = pd.DataFrame(randn(5,4), ['A','B','C','D','E'], ['W','X','Y','Z'])
df
```

```python
df > 0
```

```python
df[df>0]
```

```python
df['W'] > 0
```

```python
df[df['W']>0]
```

## Multiple Conditional Selection

```python
df[(df['W']>0) & (df['Y']<0)]
# can't use regular pythonic and here
```

```python
df[(df['W']>0) | (df['Y']<0)]
```

## Changing index

```python
df.reset_index()
# Not inplace
```
